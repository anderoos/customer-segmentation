```python
import pandas as pd
```

```python
file_path = 'customer_data.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Create a function to define age ranges
def categorize_age(age):
    if age < 18:
        return 'Under 18'
    elif 18 <= age < 25:
        return '18-24'
    elif 25 <= age < 35:
        return '25-34'
    elif 35 <= age < 45:
        return '35-44'
    elif 45 <= age < 55:
        return '45-54'
    elif 55 <= age < 65:
        return '55-64'
    else:
        return '65 and above'

# Apply the function to create a new 'Age Range' column
df['Age Range'] = df['Customer Age'].apply(categorize_age)

# Display the updated DataFrame
df.head()
```

```python
df.to_csv("customer_data_updated.csv", index=False)
```
