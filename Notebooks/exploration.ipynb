```python
# Dependencies
import pandas as pd
import numpy as np

# Viz
import matplotlib.pyplot as plt
import seaborn as sns

# Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in data
df_main = pd.read_csv('../Resources/raw/ecommerce_customer_data_custom_ratios.csv')
df_main['Purchase Date'] = pd.to_datetime(df_main['Purchase Date'])
df_main.info()
```

```python
def plot_counts(df, columns, size=(12,8)):
    fig, ax = plt.subplots(len(columns), 1, figsize=size)
    for col, axes in zip(columns, ax.flatten()):
        print(f'Plotting {col}')
        sns.countplot(df[col], ax=axes)
        axes.set_title(f'{col}')
    plt.tight_layout()
    plt.show()
    fig.savefig('../Images/EDA/cmg.png', dpi=300, format='png')
    plt.clf()
    
def plot_hist(df, columns, size=(12,8), filename='default'):
    fig, ax = plt.subplots(len(columns), 1, figsize=size)
    for col, axes in zip(columns, ax.flatten()):
        print(f'Plotting {col}')
        sns.histplot(df[col], ax=axes)
        axes.set_title(f'{col}')
    plt.tight_layout()
    plt.show()
    fig.savefig(f'../Images/EDA/{filename}.png', dpi=300, format='png')
    plt.clf()
```

```python
columns_of_interest_for_count = ['Product Category', 'Payment Method', 'Gender']
plot_counts(df_main, columns_of_interest_for_count)
```

```python
columns_of_interest_for_hist = ['Purchase Date', 'Product Price', 'Total Purchase Amount']
columns_of_interest_for_hist_2 = ['Quantity', 'Returns', 'Gender', 'Customer Age', 'Churn']

plot_hist(df_main, columns_of_interest_for_hist, size=(12,14), filename='hist1')
```

```python
plot_hist(df_main, columns_of_interest_for_hist_2, size=(12,24), filename='hist2')
```
